利用order()函數,選一種類型的資料出來,由大到小排序(decreasing = T)

```

install.packages("SportsAnalytics")

library(SportsAnalytics)
NBA1516<-fetch_NBAPlayerStatistics("15-16")
```

```
NBA1516Order<-NBA1516[order(NBA1516$TotalMinutesPlayed,decreasing=T),]
NBA1516Order[1:20,]
```

用subset()做欄位條件搜尋

```
subset(NBA1516,Team=="BOS")
```

用grepl()做字串條件搜尋

```
NBA1516[grepl("Kelly",NBA1516$Name),]
```

用ggplot畫出,四種不同類型的圖表(Exp:散佈圖、長條圖、折線圖...)

```
library(ggplot2)
df <- data.frame(gp = factor(rep(letters[1:3], each = 10)),y = rnorm(30))

ggplot(df, aes(x = gp, y = y)) +geom_point( )
```

```
ggplot(df, aes(x = gp, y = y)) +geom_boxplot( )
```

```
df$z<-df$y+rnorm(30)

#行切
ggplot(df, aes(x = z, y = y)) +geom_point( )+facet_grid(gp~.)

#直切
ggplot(df, aes(x = z, y = y)) +geom_point( )+facet_grid(.~gp)
```

```
#行切+趨勢圖
ggplot(df, aes(x = z, y = y)) +geom_point( )+facet_grid(gp~.)+geom_smooth()

#直切+趨勢圖
ggplot(df, aes(x = z, y = y)) +geom_point( )+facet_grid(.~gp)+geom_smooth()
```

```
ggplot(df, aes(x = z, y = y, color=gp)) +geom_line( )+facet_grid(gp~.)
```
